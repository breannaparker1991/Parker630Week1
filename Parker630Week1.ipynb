```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing dataset
df = pd.read_csv('fastfood.csv')
```

```python
df = df.dropna(subset=['fiber'])
```

```python
#creating a heatmap to see correlation of data
plt.figure(figsize = (16, 10))
sns.heatmap(df.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

Create a histogram or bar graph from your data.

```python
#bar graph
fig, ax = plt.subplots(figsize=(20,5) )
ax.bar(df['restaurant'], df['protein'])

ax.set_ylabel('protein')
ax.set_xlabel('restaurant')

plt.show()
```

```python
#bar graph
fig, ax = plt.subplots(figsize=(20,5) )
ax.bar(df['restaurant'], df['calories'])

ax.set_ylabel('calories')
ax.set_xlabel('restaurant')

plt.show()
```

Create a boxplot from your data.

```python
plt.figure(figsize = (10, 5))
sns.boxplot(x = 'cholesterol', y = 'total_fat', data = df)
plt.show()
```

Create a bivariate plot from your data.

```python
#scatter plot comparing inventory to sale to list ratio
fig, ax = plt.subplots(figsize=(5,5))

ax.scatter(df['fiber'], df['sugar'])
ax.set_ylabel('sugar')
ax.set_xlabel('fiber')
plt.show()
```

```python
#sorting fiber values
f = df.sort_values('fiber')
f
```

```python
#only looking at top ten fiber meals
tail = f.tail(10)
tail
```

```python
#going to see what items have the lowest calories
plt.style.use('_mpl-gallery')

# make data
x = tail['item']
y = tail['calories']

# plot
fig, ax = plt.subplots(figsize=(25,10))

ax.stem(x, y)

ax.set_ylabel('calories')
ax.set_xlabel('item')

#plt.xlim(150000, 400000)
#plt.ylim(0.9, 1.05)


plt.show()
```
